### 2. Explore variable relationships using a heatmap

```python
# Create a subset of variables for exploring relationships

sub = df[['pct_laccess', 'pct_child_laccess', 'pct_senior_laccess','pct_adult_diabetes', 'pct_adult_obese', 'med_income', 'pct_food_insec']]
```

```python
sub
```

```python
# Create a subplot with matplotlib

f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn 
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

```python
# Save figure

corr.figure.savefig("food_environ_heatmap.png") 
```

#### The heatmap shows the strength of relationships between the selected variables. 

Weak positive relationship between food insecurity and adult diabetes. Areas with higher food insecurity also have a higher percentage of adults with diabetes.

Weak negative relationships:
- between low food source access and adult diabetes: when access to food is more difficult, adult diabetes rates are higher.
- between median income and adult obesity: in areas where incomes are lower, adult obesity rates are higher.
- between food insecurity and median income: in areas where food insecurity is higher, median incomes are lower.

Moderate positive relationships:
- between food insecurity and adulty obesity: a higher rate of food insecurity in an area corresponds to a higher rate of adults who are obese.
- between adult obesity and diabetes: higher rates of obesity and diabetes are correlated. 

Moderate negative relationship between adult diabetes and median income: areas with lower incomes have higher rates of people with diabetes.

Strong positive relationships between total population low food source access and children and seniors with low food source access. Areas that have low access to food sources across the total population have high rates of children and seniors who have low access to food sources.

### 5. Examining a categorical variable

```python
# Create a histogram to visualize the distribution of the "percent with low food access" variable 

sns.histplot(df['pct_laccess'], bins = 20, kde = True)
```

```python
# Create categorical plot using histogram as a guide

df.loc[df['pct_laccess'] < 20, 'access_category'] = 'Small low access'
```

```python
df.loc[(df['pct_laccess'] >= 20) & (df['pct_laccess'] < 50), 'access_category'] = 'Medium low access'
```

```python
df.loc[df['pct_laccess'] >= 50, 'access_category'] = 'Large low access'
```

```python
df['access_category'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot to visualize the relationships between "percent of low income/ low access," "percent of low access," and "percent food insecure"

sns.set(style="ticks")
g = sns.catplot(x="pct_lowinc_laccess", y="pct_laccess", hue="access_category", data=df)
```

```python
# Save figure

g.savefig("access_cat_plot.png")
```

#### Discussion of categorical plot results

This demonstrates that the higher the number of people with low food access, the higher the number with low incomes. Despite the high variability in the data, low access to food sources does correspond to low income. However, there isn't a relationship between median income and low access to food sources. This is definitely a factor to explore further, and I plan to use the additional variable of location in further examination of the interplay of variables.
